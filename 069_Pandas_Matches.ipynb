```python
def match_city(city):
    mask = data['city']==city
    return data[mask].shape[0]
print(match_city('Hyderabad'))
print(match_city('Delhi'))
```

```python
def match_city(city):
    mask = data['city']==city
    ans=  data[mask].shape[0]
    print(f"The no of matches played in the city {city} is {ans}")

match_city('Mumbai')
match_city('Hyderabad')
match_city('Delhi')
```
