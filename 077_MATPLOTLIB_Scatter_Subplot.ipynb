# SUBPLOTS

```python
x = np.arange(1,10,.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(1,10,.1)
y = np.tan(x)
plt.plot(x,y)
plt.show()
```

```python
# plottin in the same canvas
```

```python
x = np.arange(1,10,.1)
y1 = np.sin(x)
y2 = np.tan(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
# using subplots
```

```python
x = np.arange(1,10,.1)
y1 = np.sin(x)
y2 = np.tan(x)
y3 = np.cos(x)

fig , ax = plt.subplots(1,3)

ax[0].plot(x,y1)
ax[1].plot(x,y2)
ax[2].plot(x,y3)
plt.show()
```
