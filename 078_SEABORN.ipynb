# USING subplots

```python
x = np.arange(1,10,.1)
y1 = np.sin(x)
y2 = np.tan(x)
y3 = np.cos(x)

fig , ax = plt.subplots(1,3)

ax[0].plot(x,y1)
ax[1].plot(x,y2)
ax[2].plot(x,y3)
plt.show()
```

# USING subplot

```python
x = np.arange(1,10,.1)
y1 = np.sin(x)
y2 = np.tan(x)
y3 = np.cos(x)

plt.figure() #CREATING A FIGURE OBJECT
# plt.subplot(rows,coln,seq)
plt.subplot(3,1,1)
plt.plot(x,y1)

plt.subplot(3,1,2)
plt.plot(x,y2)

plt.subplot(3,1,3)
plt.plot(x,y3)
```

```python
x = np.arange(1,10,.1)
y1 = np.sin(x)
y2 = np.tan(x)
y3 = np.cos(x)

plt.figure()
# plt.subplot(rows,coln,seq)
plt.subplot(1,3,1)
plt.plot(x,y1)

plt.subplot(1,3,2)
plt.plot(x,y2)

plt.subplot(1,3,3)
plt.plot(x,y3)
```

```python
x = np.arange(1,10,.1)
y1 = np.sin(x)
y2 = np.tan(x)
y3 = np.cos(x)

plt.figure()
# plt.subplot(rows,coln,seq)
plt.subplot(3,3,1)
plt.plot(x,y1)

plt.subplot(3,3,3)
plt.plot(x,y2)

plt.subplot(3,3,9)
plt.plot(x,y3)
```

# 3D Plotting using MATPLOTLIB

```python
a = np.arange(-1,1,0.005)
b = a

a,b = np.meshgrid(a,b)


fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(a,b,a**2+b**2)
plt.show()
```
