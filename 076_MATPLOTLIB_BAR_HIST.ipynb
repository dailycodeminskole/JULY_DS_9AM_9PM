```python
import matplotlib.pyplot as plt
import numpy as np
```

# BAR PLOT

```python
# shop1
# fruit1 :  10
# fruit2 :  20
# fruit3 :  5
```

```python
plt.bar([1,2,3],[10,20,5])
plt.show()
```

```python
plt.bar(['fruit1','fruit2','fruit3'],[10,20,5])
plt.show()
```

```python
# changin the derfault width of the bar plot
```

```python
plt.bar([1,2,3],[10,20,5],width=0.2)
plt.legend("Shop1")
plt.show()
```

```python
# shop2
# fruit1 :  5
# fruit2 : 10
# fruit3  : 6
```

```python
plt.bar([1,2,3],[5,10,6])
plt.legend("Shop2")
plt.show()
```

```python
plt.bar([1,2,3],[10,20,5])
plt.legend("Shop1")

plt.bar([1,2,3],[5,10,6])
plt.legend("Shop2")
plt.show()
```

```python
# plotting the charts side by side
plt.bar([1,2,3],[10,20,5],width=0.2)
plt.bar([1.2,2.2,3.2],[5,10,6],width=0.2)
plt.legend(["Shop1",'Shop2'])
plt.show()
```
