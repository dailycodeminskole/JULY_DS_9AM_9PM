# role of return in function

```python
def salry(x,y):
    return x+y
    
total = salry(100,5)
print(total*2)
```

```python
def salry(x,y):
    return x+y
    print("salary for jan 2025") #this will be ignored
    
total = salry(100,5)
print(total*2)
```

```python
# no code will be executed after the return line
```

```python
def salry(x,y):
    return x+y # 1st return statement will be considered
    return x-y # this will be ignored
    
total = salry(100,5)
print(total*2)
```

```python
def salry(x,y):
    return x+y ,x-y 
    
total = salry(100,5)
print(total)
print(total*2)
print(type(total))
```

# default values

```python
def calc(a,b,c):
    print(a+b+c)

# calc(10,20,30)
```

```python
# required sum of two numbers only 

def calc(a,b,c):
    print(a+b+c)

calc(10,20,0)
```

```python
# required sum of two numbers only 

def calc(a,b,c=0):
    print(a+b+c)

calc(10,20)#default value of c will be used in the additon
```

```python
# required sum of two numbers only 

def calc(a,b,c=0):
    print(a+b+c)

calc(10,20,100)
```

```python
# required sum of two numbers only 

def calc(a=0,b=0,c=0):
    print(a+b+c)

calc(10)
```

```python
# required sum of two numbers only 

def calc(a=0,b=0,c=0):
    print(a+b+c)

calc()
```
