```python
sector = company.groupby('Sector')
```

```python
# checking the size of each sector
sector.size()
```

```python
# sorting the data in ascending order by default
sector.size().sort_values()
```

```python
sector.size().sort_values(ascending=False)
```

```python
sector.first() #to get the top company in every sector
```

```python
sector.last() #to get the last company in every sector
```

```python
sector.groups #will give you the index position of the company as per the renk in every sector
```

```python
type(sector.groups)
```
