# saving a plot

```python
x = [0,1,3,5,7,9]
y = [3,9,11,12,13,17]
plt.plot(x,y) #default type is LINE plot
plt.savefig('First-image.jpg')
plt.show() #for other code editors
```

# more complex plots

```python
x_val = [-3,-2,-1,0,1,2,3]
y_val = [9,4,1,0,1,4,9]
plt.plot(x_val,y_val)
```

```python
# plotting a square fun
x_val = np.array([-3,-2,-1,0,1,2,3]) # to do elementwise operation
y_val = x_val**2
plt.plot(x_val,y_val)
```

```python
# plotting a square fun
x_val = np.arange(-4,4,.1)
y_val = x_val**2
plt.plot(x_val,y_val)
```

```python
# plotting a square fun
x_val = np.arange(-4,4,.1)
y_val = x_val**3
plt.plot(x_val,y_val)
```
