# multiple plots in the same canvas

```python
x_val = np.arange(-3,3,.5)

y1_val = x_val**2
y2_val = x_val**3
plt.xticks(np.arange(-10,10,1),rotation= 90)
plt.plot(x_val,y1_val,c='r') #sq fun
plt.plot(x_val,y2_val,c='g') # cube fun
plt.show()
```

# adding legends

```python
x_val = np.arange(-3,3,.5)

y1_val = x_val**2
y2_val = x_val**3
plt.xticks(np.arange(-10,10,1),rotation= 90)
plt.plot(x_val,y1_val,c='r',label="Square") #sq fun
plt.plot(x_val,y2_val,c='g',label="Cube") # cube fun
plt.legend()
plt.show()
```

```python
x_val = np.arange(-3,3,.5)

y1_val = x_val**2
y2_val = x_val**3
plt.xticks(np.arange(-10,10,1),rotation= 90)
plt.plot(x_val,y1_val,c='r',label="Square") #sq fun
plt.plot(x_val,y2_val,c='g',label="Cube") # cube fun
plt.legend(['Square','Cube'],loc='lower right',frameon=False,ncol=2)
plt.show()
```

```python
x_val = np.arange(-3,3,.5)

y1_val = x_val**2
y2_val = x_val**3
plt.xticks(np.arange(-10,10,1),rotation= 90)
plt.plot(x_val,y1_val,c='r',label="Square") #sq fun
plt.plot(x_val,y2_val,c='g',label="Cube") # cube fun
plt.legend(['Square','Cube'],loc='lower right',frameon=False,ncol=2)#chamgin the location and view of the legend

plt.text(0+.5,0,"One")
plt.text(1+.3,1,"Two")

plt.xlim(-2,2)
plt.ylim(-2,2)

plt.grid()# to amke the grid visible
plt.show()
```
